```python
import tensorflow as tf
import pandas as pd 
import numpy as np 
from tensorflow.contrib.layers import fully_connected  as ff
from sklearn.model_selection import train_test_split as tt
from tensorflow.contrib.layers import batch_norm
```

```python
sess=tf.InteractiveSession()
```

```python

df=pd.read_csv('data.csv')
df=df.drop(['id','Unnamed: 32'],axis=1)
XX=np.array(df.drop(['diagnosis'],axis=1)).astype(np.float32)
yy=np.array((df['diagnosis']=='M').astype(np.int)).astype(np.int64)
Xt,Xl,yt,yl=tt(XX,yy,train_size=0.2)
df.head()
```

```python
X=tf.placeholder(tf.float32,shape=(None,XX.shape[1]))
y=tf.placeholder(tf.int64,shape=(None))
is_training = tf.placeholder(tf.bool, shape=())
bn_params = {
'is_training': is_training,
'decay': 0.99,
'updates_collections': None
}
hidden1=ff(X,15,activation_fn=tf.nn.relu,normalizer_fn=batch_norm, normalizer_params=bn_params)
hidden2=ff(hidden1,10,activation_fn=tf.nn.relu,normalizer_fn=batch_norm, normalizer_params=bn_params)
logits=ff(hidden2,2,activation_fn=None,normalizer_fn=batch_norm, normalizer_params=bn_params)

xentropy=tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)
loss = tf.reduce_mean(xentropy)

learning_rate = 0.01
optimizer = tf.train.AdamOptimizer(learning_rate)
training_op = optimizer.minimize(loss)

correct = tf.nn.in_top_k(logits, y, 1)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

init=tf.global_variables_initializer()
```

```python
feed={X:Xt,y:yt,is_training:False}
deef={X:Xl,y:yl,is_training:True}
sess.run(init)
for i in range(100):
    sess.run(training_op,feed_dict=deef)
    print(accuracy.eval(feed_dict=feed),accuracy.eval(feed_dict=feed))
```
